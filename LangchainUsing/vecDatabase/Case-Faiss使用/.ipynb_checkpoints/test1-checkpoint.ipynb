```python
import pandas as pd

# 数据加载
news = pd.read_csv('./text_similarity/sqlResult.csv', encoding='gb18030')
news
```

```python
# 89611篇新闻
news.shape
```

```python
news['content'].isnull().sum()
news.dropna(subset=['content'], inplace=True)
```

```python
import jieba
# 87054
news.shape
# 对文本进行分词
def split_text(text):
    text = text.replace(' ', '')
    text = text.replace('\n', '').replace('\r', '')
    text2 = jieba.cut(text)
    result = ' '.join([x for x in text2])
    return result
# 分词
split_text(news.iloc[0].content)
```

```python
# 对所有文本进行分词
corpus = list(map(split_text, [str(x) for x in news['content']]))
corpus[0]
```

```python
import pickle
with open('./corpus.pkl', 'wb') as file:
    pickle.dump(corpus, file)
"""
with open('./corpus.pkl', 'rb') as file:
    corpus = pickle.load(file)
"""
```

```python
# 计算corpus的 TFIDF
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
countvectorizer = CountVectorizer(encoding='gb18030', min_df=0.01)
tfidftransformer = TfidfTransformer()

# 得到单词数量的vector
countvector = countvectorizer.fit_transform(corpus)
countvector.shape
```

```python
tfidf = tfidftransformer.fit_transform(countvector)
tfidf.shape
```

```python
# 假设你是新华社，标记是否为自己的新闻
label = list(map(lambda source: 1 if '新华社' in str(source) else 0, news['source']))
#label
news.iloc[3]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

import lightgbm as lgb
# 训练一个预测器：是否为新华社风格的文章
x_train, x_test, y_train, y_test = train_test_split(tfidf.toarray(), label, test_size=0.3, random_state=2022)
model = MultinomialNB()
"""
model = lgb.LGBMClassifier(
            num_leaves=2**5-1, reg_alpha=0.25, reg_lambda=0.25, objective='binary',
            max_depth=-1, learning_rate=0.05, min_child_samples=3, random_state=2022,
            n_estimators=100, subsample=1, colsample_bytree=1,
        )
"""
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
# 模型预测
y_pred = model.predict(x_test)
print('accuracy: ', accuracy_score(y_test, y_pred))
print('precision: ', precision_score(y_test, y_pred))
print('reclass: ', recall_score(y_test, y_pred))
```

```python
import numpy as np

# 使用模型来对所有新闻进行预测
prediction = model.predict(tfidf.toarray())
labels = np.array(label)

# compare_news_index有两列： prediction预测结果，labels实际值
compare_news_index = pd.DataFrame({'prediction': prediction, 'labels': labels})
compare_news_index
```

```python
# copy_news_index: 可能是Copy的新闻（如果预测结果为1，但实际label=0）
copy_news_index = compare_news_index[(compare_news_index['prediction']==1) & (compare_news_index['labels']==0)].index
print('可能为Copy的新闻条数：', len(copy_news_index))
```

```python
# 使用KMeans对文章进行聚类
from sklearn.cluster import KMeans
from sklearn.preprocessing import Normalizer
# 数据归一化
normalizer = Normalizer()
scaled_array = normalizer.fit_transform(tfidf.toarray())
# 使用KMeans
kmeans = KMeans(n_clusters=50, random_state=2022)
k_labels = kmeans.fit_predict(scaled_array)
# 保存聚类结果
with open('k_labels.pkl', 'wb') as file:
    pickle.dump(k_labels, file)
```

```python
# 创建id_class
id_class = {}
for index, class_ in enumerate(k_labels):
    id_class[index] =class_
id_class
```

```python
# 实际为新华社的新闻
xinhuashe_news_index = compare_news_index[compare_news_index['labels']==1].index
```

```python
with open('./tfidf.pkl', 'wb') as file:
    pickle.dump(tfidf.toarray(), file)
```

```python
from collections import defaultdict

# 创建class_id
class_id = defaultdict(set)
for index, class_ in id_class.items():
    # 只统计新华社发布的class_id
    if index in xinhuashe_news_index.tolist():
        class_id[class_].add(index)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# Step1，分类
# Step2，聚类
# Step3，相似度计算

# 查找相似文本（先用聚类label结果进行过滤）
# 对某篇文章cpindex，查找新华社里面最相似的文章
def find_simlar_text(cpindex, top=10):
    # 先看cpindex是哪个类别
    temp_label = id_class[cpindex]
    # 再看这个类别里面 新华社有哪些文章
    ids = class_id[temp_label]
    dist_dict = {}
    for x in ids:
        dist_dict[x] = cosine_similarity(tfidf[cpindex], tfidf[x])
    # 从大到小进行排序
    result = sorted(dist_dict.items(), key=lambda x: x[1][0], reverse=True)[:top]
    return result
cpindex = 3352
news.iloc[cpindex]['content']
find_simlar_text(cpindex)
```
